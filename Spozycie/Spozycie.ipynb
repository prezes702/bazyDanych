```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('spozycie.csv');
df[df['beer_servings']>df['beer_servings'].mean()]

df['beer'] = df['beer_servings']/df['beer_servings'].mean()
df['spirit'] = df['spirit_servings']/df['spirit_servings'].mean()
df['wine'] = df['wine_servings']/df['wine_servings'].mean()
df['total'] = df['total_litres_of_pure_alcohol']/df['total_litres_of_pure_alcohol'].mean()

plt.rcParams["figure.figsize"] = (40,20)


plt.plot(range(len(df)), df['beer']-df['total'], '-', label='Berr')
plt.plot(range(len(df)), df['spirit']-df['total'], '-', label='Spirit')
plt.plot(range(len(df)), df['wine']-df['total'], '-', label='Wine')
plt.legend(loc='best')
plt.show()


```

```python
#określenie korelacji między spożyciem danego typu alkoholu a całkowitym spożyciu
df['beer'].corr(df['total'])
```
